```python
def func(my,xx,name_of_folder):
    text=[]
    no_of_words=[]
    publisher=[]
    date=[]
    tckr=[]
    name=[]
    headline=[]
    q=0
    while(q<len(my)):
        check=my[q].findAll("div")
        head=my[q].findAll("span",{"class":"enHeadline"})
        headline.append(head[0].text)
        te=my[q].findAll("p",{"class":"articleParagraph enarticleParagraph"})
        l=[]
        x=129029121
        for i in range(len(te)):
            l.append(te[i].text)
        text.append(" ".join(l))
        for i in range(len(check)):
            if "words" in check[len(check)-i-1].text:
                x=len(check)-i-1
                no_of_words.append(check[x].text)
                break
        if(x!=129029121):
            date.append(check[x+1].text)
            publisher.append(check[x+2].text)
        else:
            date.append("")
            publisher.append("")
        tckr.append(xx)
        name.append(name_of_folder)
        q+=2
    return text, no_of_words, publisher, date, tckr, name, headline
```

```python
#main function
t=[]
n=[]
p=[]
d=[]
h=[]
ticker=[]
company=[]
for i in glob.glob("..\\Harsh/*"):
    print(i)
    print("")
    name_of_folder=i.split('\\')[-1]
    if(name_of_folder!="5073 GENERAL MOTORS CO"):
        print(name_of_folder)
        j="..\\Harsh/"+name_of_folder+"/*.html"
        x=name_of_folder.split()[0]
        for k in glob.glob(j): 
            print(k)
            bs=BeautifulSoup(open(k, encoding="utf8"),features="lxml")
            my=bs.findAll("div",{"class": "article"})
            txt, now, pblshr, dt, tckr, name,headline= func(my,x,name_of_folder)
            t.extend(txt)
            n.extend(now)
            p.extend(pblshr)
            d.extend(dt)
            ticker.extend(tckr)
            company.extend(name)
            h.extend(headline)
```
